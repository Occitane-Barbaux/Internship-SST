```python

### The full run takes 10-15 minutes on my computer
import sys,os
sys.version
```

```python
#Paramètres

st_n="34154001" ##Used in data preparation to get the right observation file .nc
Type_station="Oui" #Different treatment depending on origin

Y_ref="MAUGUIO" # To create folders
lat=43.5761111
lon=3.9647222222222225 ## Coordinates used to select the right GCM series

ref_deb=1986
ref_fin=2016 ## Give boudaries of the reference period for anomaly !!! Has to be included in the observation period

X_ref="Europe" #Choice of covariate


absolute_path=os.path.abspath(os.getcwd())
work_path=absolute_path+"/Results/"+Y_ref
print(work_path)

observations_brutes_path=os.path.abspath(os.getcwd())+"/Example/Data_Brutes"

path_X_Modeles_Brut=observations_brutes_path+"/CMIP6_TasMean_"+X_ref
path_Y_Modeles_Brut=observations_brutes_path+"/CMIP6_TXX_France_Full"

path_X_Obs_Brut=observations_brutes_path+"/ERA5_Obs_"+X_ref+"/ERA5_tas_"+X_ref+".nc"
path_Y_Obs_Brut=observations_brutes_path+"/Observations_Stations/"

### !! Beware: The code is **very** sensitive to spaces in paths
```

```python
%%time
### Data prepararation
## This part really depend on your entry files !
## Probably the file with the most modificatiosn to do for you
!python3 Utilities/Data_Preparation.py $Y_ref $X_ref $work_path $path_X_Modeles_Brut $path_Y_Modeles_Brut $path_X_Obs_Brut $path_Y_Obs_Brut $lat $lon $ref_deb $ref_fin $Type_station $st_n

print("Data preparation done")

## !! In case of weird error message / cdo related on second run, try deleting coresponding temporary files in results/[yourplace]/data
```

```python
%%time
### Calculations
## Rely on "Functions_run_Model.py" for functions used
## There's likely references to tas and tasmax you'll have to change yourself in both "Functions_run_Model.py" and "Full_run_Model.py"
## !!!! Do not forget to change the path of "set_cmdstan_path" in "Full_run_Model.py" line 10.

pathObs=work_path+"/data/Observations/03_Post_Treatment/"+Y_ref+"_"+X_ref
pathGCM_X_separated=work_path+"/data/CMIP6/03_Post_Treatment/TasMean_"+X_ref+"/separated"
pathGCM_Y_separated=work_path+"/data/CMIP6/03_Post_Treatment/TXX_"+Y_ref+"/separated"

results =!python3 Utilities/Full_run_Model.py $Y_ref $X_ref $work_path $pathObs $pathGCM_X_separated $pathGCM_Y_separated $ref_deb $ref_fin
dt_string=results[-1]
print(results)
print(dt_string) ## This gives you a (mostly) unique reference number for each run of the method, in case you run it several times. Used for figures too
print("Calculations done")
```
